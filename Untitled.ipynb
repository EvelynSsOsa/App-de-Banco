```python
##metdodo de clase  
class Pajaro: 

    alas = True ## atributo de clase 

    def __init__(self,color,especie): 
        self.color = color ## atributo de instancia de clase 
        self.especie = especie  
        
    def piar(self): ## aqui tenemos una función dentro de una clase   
        print('pio, mi color es {}'.format(self.color)) ## con ".format(self.color)" -> traemos el atibuto de instancia 

    def volar(self,metro): ## colocamos un parametro 
        print(f'el pajaro a volado {metro} metros')## para mandarlo a llamar en un mensaje    
        self.piar()#madamos a llamar a otro métdodo, de esta forma accedemos a otro metodo dentro de este mismo objeto ##2.-

        
##"pintar_negro()" -> es un método de instancia que puede acceder y modificar atributos del objeto
    def pintar_negro(self):
        self.color = 'negro'## en este caso accedio a "self.color" y lo modifico  a negro ##1.-
        print(f'Ahora el pajaro es color {self.color}') 

## Aquí tenemos un metodo de clase, le pasas como paramtero un decorador de clase "cls",
    ##que nos indica que esta NO esta relaciónada con la clase en si y no con las intancias 
    @classmethod 
    def poner_huevos(cls,cantidad):## 
        print(f'Puso {cantidad} huevos')  
        # print(f'es de color [self.color]') -> no podemos recibir atributos de instancia porque este metdodo solo recibe cls y no self 
        cls.alas = False ## sin embargo en los metdodos de clase si podemos modificar atributos de clase 
        print(Pajaro.alas) 


    @staticmethod
    def mirar(): 
        print("El parajo mira")

## Aquí no es necesario pasar una intancias como la tenia antes con "PIOLIN"
  ##EJEMPLO:      #piolin = Pajaro('amarillo','canario')
## en su lugar se hace así: Pajaro.poner_huevos(3) 
Pajaro.mirar()
```

```python
## entonces ya sabemos: 
metodos de clase: 
pueden modificar y recibir atributos de clase (son la variables que dentro de la clase, no van dentro de ningún metodo)
pueden recibir parametros "cls" 

metodos de instancia: 
pueden modificar y recibir atributos de intancia (que son las variables que se van declarando dentro de la funciónes) 
pueden recibir parametros como "self"  

metodos estaticos:
No pueden modificar estados de una instancia, ni tampoco se pueden cambiar los atributos de una instancia 
No se accede ni atributos de intancia, ni atributos de clase... 
Nos srive para que no sean cambiados o modificados los atributos o funciónes que realicemos 
```

```python
##Polimorfismo: 

class Vaca: 

    def __init__(self,nombre): ## creamos atributos de instancia, solo ocupamos nombre
        self.nombre = nombre 

    def hablar(self): 
        print(self.nombre + " dice muuuuu🐮🐮") ### creamos una clase de una vaquita que hace "Muuuuu 🐮🐮🐮"


class Oveja: 

    def __init__(self,nombre): 
        self.nombre = nombre ## creamos atributos de instancia 

    def hablar(self): 
        print(self.nombre + " dice be be 🐑🐑") ### creamos una clase de ovj


vaca_uno = Vaca("Lulu") ## en la instancia "vaca_uno" se llama a la clase "Vaca" y pasamos como paramtero el nombre que tendra nuestra vaquita
oveja_uno = Oveja("Cheep") ## en la instancia "oveja_uno" se llama a la clase "Oveja" y pasamos como parametro el nombre de nuestra ovejita 

#######AQUI ES DONDE ENTRA EL POLIMORFISMO: -> Es decir que creamos una función, que tendra una funcion en comun que tengan nuestros objetos
def animales(animal):## se pasa un paremtro -> en este caso ese parametro sera un animalito creado 
    animal.hablar() ## madamos a llamar a una funcion "hablar"...antes ya fue creada... que concatena el nombre del animalito y su frase

animales(oveja_uno)  ## con la func "animales" pasamos como parametro "oveja_uno" -> nos devuelve "Cheep dice be be 🐑🐑"
animales(vaca_uno) 
    

    
```

```python
class Mago():
    def defender(self):
        print("Escudo mágico")

class Arquero():
    def defender(self):
        print("Esconderse")

class Samurai():
    def defender(self):
        print("Bloqueo") 
        
        
        
def personaje_defender(defensas): 
    for personaje in defensas:
        personaje.defender()  
        
defensas = [Mago(),Arquero(),Samurai()]

personaje_defender(defensas)
```

```python
class CD: 

    def __init__(self,autor,titulo,canciones): 
        self.autor = autor
        self.titulo = titulo 
        self.canciones = canciones 


    def __str__(self): ## este metodo nos ayuda a poder imprimir mensajes -> "def __str__(self):"
        return f' El album {self.titulo} de {self.autor} tiene {self.canciones} canciones'

    def __len__(self): ## este metodo nos ayuda a poder contar los elementos de un objeto, es como un "len(xxx)" 
        return self.canciones  ## definimos al atributo de instancia "self.canciones"

    def __del__(self): ## este metodo nos ayuda a eliminar... en este caso lo vamos a aplicar al borrar intancias 
        print('se ha elimando el cd con exito')## y vamos a imprimir este de aquí  
        
## Creamos intancias "mi_cd" y "mi_cd_dos", que mandan a llamar a la clase "CD" ya solo pasamos oa


mi_cd = CD('Tama Impala','Deadbeat',10)   
mi_cd_dos = CD('Sabrina Carpenter','Mans Best Friend',12)  

print(mi_cd)## madamos a imprimir "mi_cd" -> El album Deadbeat de Tama Impala tiene 10 canciones 
print(len(mi_cd)) ##
del mi_cd
print(mi_cd_dos) 
print(len(mi_cd_dos)) 
## si imprimimos este de nuevo "print(mi_cd)" -> nos devolvería error, porque ya fue eliminado con "del mi_cd"
```

```python
###Ejercicio: 
## Dada la clase Libro, implementa el método especial __str__ para que cada vez que se imprima el objeto, devuelva '"{titulo}", de {autor}' 
##(atención: el título debe estar encerrado entre comillas dobles).

class Libro():
    def __init__(self, titulo, autor, cantidad_paginas):
        self.titulo = titulo
        self.autor = autor
        self.cantidad_paginas = cantidad_paginas## aqui tenemos los atributos de instancia 
        
    def __str__(self): ## este metodo nos va ayudar a imprimir en pantalla los atributos de instancia 
        return f'"{self.titulo}", de {self.autor} con un total de {self.cantidad_paginas} paginas📗' 
        ## el ejercicio no lo pide, pero aqui le hice una modificación, para también poder ver el número de paginas

## creamos la instancia, donde madamos a llamar a la clase y pasamos los parametros
mi_libro = Libro("El alquimista","Paulo Coelho",200) 
print(mi_libro) ## imprimimos y mostramos en pantalla la instancia 
```
